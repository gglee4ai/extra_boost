```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
t = np.linspace(0, 1, 1000)
dt = t[1] - t[0]
```

```python
f = 3 * np.sin(12.3*t) + 4 * np.cos(12.3*t) + np.sin(56.7*t) + np.cos(56.7*t)
```

```python
fig = plt.figure(figsize=(16, 6))
plt.plot(t, f)
plt.show()
plt.close(fig)
```

```python
X = np.concatenate((f[0:-4].reshape(-1, 1), 
                    f[1:-3].reshape(-1, 1),
                    f[2:-2].reshape(-1, 1), 
                    f[3:-1].reshape(-1, 1)), axis=1)
y =                 f[4:  ].reshape(-1, 1)
```

```python
X.shape, y.shape
```

```python
a = (X.T.dot(X))
```

```python
pv = np.linalg.inv(a).dot(X.T.dot(y))
```

```python
pv.T.squeeze()
```

```python
kappa = np.roots(np.concatenate((np.array([-1]), pv.T.squeeze())))
kappa
```

```python
omega = kappa / dt
```
