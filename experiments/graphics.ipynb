```python
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
#rc('font',**{'family':'serif','serif':['CMU Serif']})

def drawAUC(qs, qnames, figname):
    f = plt.figure(figsize=(5.5,3))
    plt.rc('text', usetex=True)
    plt.xlabel(r'$t$, time feature')
    plt.ylabel(r'AUC',fontsize=16)
    tt_new1 = tt_new[tt_new > 0]
    plt.plot(tt_new1, 0.5*(tt_new1==tt_new1), '--', color="#bbbbbb", label='neutral score')
    for qq, name in zip(qs, qnames):
        plt.plot(tt_new,qq,label='\\large \\strut ' + name)    
    #plt.title(r"\TeX\ is Number "
    #          r"$\displaystyle\sum_{n=1}^\infty\frac{-e^{i\pi}}{2^n}$!",
    #          fontsize=20, color='gray')
    xt = np.linspace(0, 1, 11)
    plt.xticks(xt, ["${:.2}$".format(xx) for xx in xt])

    plt.legend(loc='lower left')
    plt.savefig(figname, dpi=300)
    plt.show()
    
    plt.close(f)
    
if 1:
    qnames = ['Extra Boost', 'Regular', 'XGBoost, $d=1$', 'XGBoost, $d=4$', 'Extra .5', 'Regular .5', 'XGBoost 1 .5', 'XGBoost 4 .5']
    drawAUC(qs[:4], qnames[:4], 'fig.pdf')
```
